```python
import re

text = text.replace('\n', '')
sentence = re.split('\. |\.',text)
sentence
```

```python
len(sentence)
```

#2. Preprocessing

##2.1 Tokenize

```python
import nltk
nltk.download('punkt')
```

```python
tokenizer = nltk.RegexpTokenizer(r"\w+")
tokenized = [tokenizer.tokenize(s.lower()) for s in sentence]
tokenized
```

##2.2 Stopword Removal

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
listStopword =  set(stopwords.words('indonesian'))
listStopword
```

```python
important_token = []
for sent in tokenized:
  filtered = [s for s in sent if s not in listStopword]
  important_token.append(filtered)
important_token
```

```python
sw_removed = [' '.join(t) for t in important_token]
sw_removed
```

##2.3 Stemming

```python
!pip install Sastrawi
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
stemmed_sent = [stemmer.stem(sent) for sent in sw_removed]
stemmed_sent
```

#3. Extract Feature

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vec = TfidfVectorizer(lowercase=True)
document = vec.fit_transform(stemmed_sent)
```

```python
document = document.toarray()
```

```python
document.shape
```

#4. Choose Top-n sentence

```python
n = 5 #pilih berapa kalimat sebagai hasil ringkasan
```

```python
import numpy as np
```

```python
result = np.sum(document, axis=1)
result.shape
```

```python
sorted(result)
```

```python
top_n = np.argsort(result)[-n:]
top_n
```

```python
summ_index = sorted(top_n)
summ_index
```

```python
print('Hasil Ringkasan: ')
for i in summ_index:
  print(sentence[i], end='.\n')
```
